```python
import pandas as pd

df=pd.read_csv('../app/data/jobs.csv')

df.head()
```

```python
features = ["Job Title", "Job Description"]

for feature in features:
    df[feature] = df[feature].str.lower()
```

```python
import pandas as pd
import re  

def parse_salary(salary):
    # Supprimer tout sauf chiffres, tiret et point
    salary = re.sub(r'[^\d\-.]', '', salary)
    
    if '-' in salary:
        low, high = salary.split('-')
        return (float(low.strip()) + float(high.strip())) / 2
    else:
        return float(salary.strip())

# Appliquer à toute la colonne
df['Salary_Estimate'] = df['Salary Estimate'].apply(parse_salary)
print(df[['Salary_Estimate']])

```

```python
from sklearn.preprocessing import OrdinalEncoder

most_freq = df['Revenue'].value_counts()

second_most_freq = most_freq.index[1]

df['Revenue'] = df['Revenue'].replace({
    "Unknown / Non-Applicable": second_most_freq,
    '-1' :second_most_freq
})


revenue= [
'Less than $1 million (USD)'    ,
'$1 to $5 million (USD)'          ,
'$5 to $10 million (USD)'     ,
'$10 to $25 million (USD)'     ,
'$25 to $50 million (USD)'   ,
'$50 to $100 million (USD)'         ,
'$100 to $500 million (USD)'     ,
'$500 million to $1 billion (USD)' ,
'$1 to $2 billion (USD)'    ,
'$2 to $5 billion (USD)'        ,
'$5 to $10 billion (USD)'     ,
'$10+ billion (USD)'              ,
]

encoder=OrdinalEncoder(categories=[revenue])

df['Revenue']=encoder.fit_transform(df[['Revenue']])
print(df['Revenue'])
```

```python
most_freq = df['Size'].mode()[0]

df['Size'] = df['Size'].replace(["Unknown", "-1"], most_freq)


from sklearn.preprocessing import OrdinalEncoder

size= [
'1 to 50 employees'  ,
'51 to 200 employees',
'201 to 500 employees'  ,
'501 to 1000 employees' ,
'1001 to 5000 employees' ,
'5001 to 10000 employees',
'10000+ employees',
]

encoder=OrdinalEncoder(categories=[size])

df['Size']=encoder.fit_transform(df[['Size']])
print(df['Size'])
```

```python
from sklearn.preprocessing import LabelEncoder

lencoder=LabelEncoder()
df['Location']=lencoder.fit_transform(df['Location'])
df['Company_Name']=lencoder.fit_transform(df['Company Name'])
```

```python
most_freq=df["Type of ownership"].value_counts().idxmax()

df['Type of ownership'] = df['Type of ownership'].replace({
   "Unknown": most_freq,
   "-1":most_freq
})
```

```python
df['Type_ownership']=lencoder.fit_transform(df['Type of ownership'])
```

```python
most_freq = df['Industry'].value_counts()
second_most_freq = most_freq.index[1]

df['Industry']=lencoder.fit_transform(df['Industry'])

```

```python

df["Sector"].value_counts()
most_freq=df["Sector"].value_counts().idxmax()

df['Sector'].replace({
   "Unknown": most_freq,
   "-1":most_freq
})

df['Sector']=lencoder.fit_transform(df['Sector'])
```

```python
df['Competitors'] = df['Competitors'].apply(lambda x: 0 if x == '-1' else 1)
```

```python
df['Job_Title']=df['Job Title']
df['Job_Description']=df['Job Description']
```

```python
import re

def clean_text(text):
    # Convert to string (in case of NaN)
    text = str(text)
    
    # Lowercase
    text = text.lower()
    
    # Remove \n and extra spaces
    text = re.sub(r'\n+', ' ', text)
    
    # Remove special characters (keep letters and numbers)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df['Job_Description'] = df['Job_Description'].apply(clean_text)
```
